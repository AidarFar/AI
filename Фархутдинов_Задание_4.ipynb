Входные данные:
1.	Связный граф не менее 50 вершин.
2.	Веса ребер в диапазоне от 1 до 10 случайным образом.


Задача 1.
1.	Построить остовное дерево минимального веса на графе, используя алгоритм Краскала.
2.	Визуализировать решение.


```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Создаем связный граф
G = nx.connected_watts_strogatz_graph(50, 4, 0.3)

# Генерируем случайные веса для ребер
for u, v in G.edges():
    G[u][v]['weight'] = random.randint(1, 10)

# Визуализируем граф
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_size=200, node_color='skyblue', font_size=10, font_weight='bold')

# Выводим веса ребер
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels,font_color="purple")

plt.title("Связный граф с случайными весами ребер")
plt.show()
```

```python
min_spanning_tree = nx.minimum_spanning_tree(G)
plt.figure(figsize=(10, 6))
pos = nx.spring_layout(G, seed=42)
nx.draw_networkx_edges(G, pos, alpha=0.4)
nx.draw_networkx_nodes(G, pos, node_size=300, node_color='skyblue')
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')
nx.draw_networkx_edges(min_spanning_tree, pos, edge_color='red', width=2)
plt.title("Остовное дерево минимального веса (алгоритм Краскала)")
plt.show()
# Вычисляем общий вес остовного дерева минимального веса
total_weight = sum([attr['weight'] for u, v, attr in min_spanning_tree.edges(data=True)])

# Выводим общий вес
print("Общий вес остовного дерева минимального веса:", total_weight)
```

1.	Провести 30 итераций задачи 2.
2.	Построить график изменения количественного и структурного изменения.


```python
import copy
import random
import matplotlib.pyplot as plt

# Создаем пустые списки для сохранения данных
total_weights_before = []
total_weights_after = []

# Проводим 30 итераций
for i in range(30):
    # Создаем копию исходного графа
    G_modified = copy.deepcopy(G)

    # Выбираем случайные 5 ребер
    edges_to_change = random.sample(G_modified.edges(), 5)

    # Изменяем веса выбранных ребер в диапазоне от 1 до 3
    for edge in edges_to_change:
        G_modified[edge[0]][edge[1]]['weight'] = random.randint(1, 3)

    # Строим остовное дерево минимального веса до изменений
    min_spanning_tree_before = nx.minimum_spanning_tree(G)
    total_weight_before = sum([attr['weight'] for u, v, attr in min_spanning_tree_before.edges(data=True)])

    # Строим остовное дерево минимального веса после изменений
    min_spanning_tree_after = nx.minimum_spanning_tree(G_modified)
    total_weight_after = sum([attr['weight'] for u, v, attr in min_spanning_tree_after.edges(data=True)])

    # Сохраняем общие веса до и после изменений
    total_weights_before.append(total_weight_before)
    total_weights_after.append(total_weight_after)

# Построение графика изменения количественного и структурного изменения
plt.figure(figsize=(10, 6))
plt.plot(total_weights_before, label='До изменений', marker='o')
plt.plot(total_weights_after, label='После изменений', marker='s')
plt.xlabel('Итерация')
plt.ylabel('Общий вес остовного дерева минимального веса')
plt.title('Изменение общего веса остовного дерева минимального веса')
plt.legend()
plt.grid(True)
plt.show()
```
