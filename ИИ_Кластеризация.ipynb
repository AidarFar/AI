```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
def self_similarity_graph(n, G=None):
    if G is None:
        G = nx.Graph()

    if n == 0:
        G.add_nodes_from([1, 2, 3])
        G.add_edges_from([(1, 2), (2, 3), (3, 1)])
        return G

    previous_graph = self_similarity_graph(n - 1, G)
    prev_max_node = max(previous_graph.nodes)
    new_nodes = [node + prev_max_node for node in previous_graph.nodes]
    new_nodes.extend([node + prev_max_node for node in new_nodes])
    G.add_nodes_from(new_nodes)

    for u in list(previous_graph.nodes)[:prev_max_node]:
        u1 = new_nodes[0]
        u2 = new_nodes[1]
        G.add_edge(u, u1)
        G.add_edge(u, u2)
        G.add_edge(u2, u1)
        new_nodes = new_nodes[2:]
    return G

for i in range(0, 6):

    G = nx.Graph()

    n = i

    self_similarity_graph(n, G)

    nx.draw(G, with_labels=True, node_color='skyblue')
    plt.show()

    adj_matrix = nx.to_numpy_array(G)

    edge_list = list(G.edges())

    df_adj_matrix = pd.DataFrame(adj_matrix.astype(int), columns=range(1, len(G.nodes()) + 1), index=range(1, len(G.nodes()) + 1))

    print("Матрица смежности:")
    print(df_adj_matrix)

    print("\nСписок рёбер:")
    print(edge_list)
```
